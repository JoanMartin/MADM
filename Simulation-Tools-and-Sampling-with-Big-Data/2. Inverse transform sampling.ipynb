```python
import random

import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12, 10)
```

```python
def plot(X, Y):
    nbins = 500
    y_1, x_1 = np.histogram(X, normed=True, bins=nbins)
    y_2, x_2 = np.histogram(Y, normed=True, bins=nbins)

    fig, ax = plt.subplots(nrows=1, ncols=2)
    ax[0].plot(x_1[:nbins], y_1)
    ax[0].set_title('Simulated')
    ax[1].plot(x_2[:nbins], y_2)
    ax[1].set_title('Original')
    plt.show()  
```

# Exponential Distribution

$$ F(x, \lambda) = 1 - e^{-\lambda x} $$

$$ 1 - U = e^{-\lambda x} $$

$$ ln(1 - U) = -\lambda x $$

$$ x = -\frac{1}{\lambda} ln(1 - U) = -\frac{1}{\lambda} ln(U) $$

```python
N = 10**6
beta = 2

U = np.random.uniform(0, 1, size = N)
X = -(1/beta) * np.log(U)
Y = np.random.exponential(1/beta, size=N)

plot(X, Y)
```

# Logistic Distribution

$$ F(x; \mu, \sigma) = \frac{1}{1 + e^{-(x - \mu) / \sigma}} $$

$$ e^{-(x - \mu) / \sigma} = \frac{1}{U} - 1 $$

$$ \frac{-(x - \mu)}{\sigma} = \ln{(\frac{1}{U} - 1)} $$

$$ -x + \mu = \sigma \cdot \ln{(\frac{1}{U} - 1)} $$

$$ x = -\sigma \cdot \ln{(\frac{1}{U} - 1)} + \mu $$

```python
N = 10**7
mu = 0
sigma = 1

U = np.random.uniform(0, 1, size = N)
X = -sigma * np.log(np.divide(1, U) - 1) + mu
Y = np.random.logistic(loc=mu, scale=sigma, size=N)

plot(X, Y)
```

# Weibull Distribution

$$ F(x, \alpha, \beta) = 1 - e^{-(x/\alpha)^\beta} $$

$$ 1 - U = e^{-(x/\alpha)^\beta} $$

$$ ln(1 - U) = -\left(\frac{x}{\alpha}\right)^\beta $$

$$ [-ln(1 - U)]^{1/\beta} = \frac{x}{\alpha} $$

$$ x = \alpha[-ln(1 - U)]^{1/\beta} = \alpha[-ln(U)]^{1/\beta} $$

```python
N = 10**7
alpha = 5
beta = 2

U = np.random.uniform(0, 1, size = N)
X = alpha * (-np.log(U))**(1/beta)
Y = alpha * np.random.weibull(beta, size=N)

plot(X, Y)
```
