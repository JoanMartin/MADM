```python
from math import sqrt
import random

import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 10)
```

# Calculando $\pi$ con gotas de lluvia

```python
def montecarlo_pi(x_c, y_c, r, k, save_points=False):   
    n_inside = 0
    points = []
    
    for i in range(1, k + 1):
        x_p = random.uniform(-1,1)
        y_p = random.uniform(-1,1)
        if save_points:
            points.append([x_p, y_p])
            
        d = sqrt((x_p - x_c)**2 + (y_p - y_c)**2)

        if d < r:
            n_inside = n_inside + 1

    p = n_inside / k
    pi = 4 * p
    
    if save_points:
        return (pi, points)
    
    return pi
```

```python
r = 1
x_c = 0
y_c = 0
k = 10000

pi, points = montecarlo_pi(x_c, y_c, r, k, True)

x, y = np.array(points).T

fig, ax = plt.subplots()
ax.add_artist(plt.Circle((x_c, y_c), r, color='r', fill=False))
ax.scatter(x, y, s=2)
ax.set_title("pi = %0.5f" % pi)
```

```python
r = 1
x_c = 0
y_c = 0
r1 = [i for i in range(1, 10000, 10)]
r2 = [i for i in range(10000, 1000000, 100000)]
k = r1 + r2
    
pi = [montecarlo_pi(x_c, y_c, r, i) for i in k]

plt.plot(k, pi)
```
